# 12. Final Prediction Pipeline

## 12.1 Import Libraries

In this section, we import all the required libraries for building the final prediction pipeline. These libraries are used to load the saved preprocessing pipeline and trained machine learning model, process new customer data, generate credit default predictions, and export the prediction results.

Using a consistent set of libraries ensures that the inference process follows the same workflow as the training process, enabling reliable and reproducible predictions on unseen customer applications.

````python
# ============================================================
# 12.1 Import Libraries
# ============================================================

# Data manipulation
import pandas as pd
import numpy as np

# Model persistence
import joblib

# File and path handling
from pathlib import Path

# Display settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

print("Libraries imported successfully.")
````

## 12.2 Load Saved LightGBM Pipeline

In this section, the saved LightGBM prediction pipeline is loaded. The pipeline contains both the fitted preprocessing component and the trained LightGBM classifier, enabling end-to-end prediction without retraining or manually applying preprocessing steps.

Using the saved pipeline ensures that new customer applications are processed using the exact same transformations and model configuration employed during training, resulting in consistent and reliable credit risk predictions.

````python
# ============================================================
# 12.2 Load Saved LightGBM Pipeline
# ============================================================

import joblib

# Load the complete LightGBM pipeline
lightgbm_pipeline = joblib.load("../models/lightgbm.pkl")

print("=" * 60)
print("LightGBM pipeline loaded successfully!")
print("=" * 60)
````

# 12.3 Load Engineered Test Dataset

````python
test_data = pd.read_csv("../data/processed/application_train_engineered.csv")
````

# 12.4  Generate Credit Default Predictions

````python
predictions = lightgbm_pipeline.predict(test_data)

default_probabilities = lightgbm_pipeline.predict_proba(test_data)[:, 1]
````

````python
print(type(predictions))
print(predictions.shape)
print(default_probabilities.shape)
````

````python
prediction_results = pd.DataFrame({
    "SK_ID_CURR": test_data["SK_ID_CURR"],
    "Predicted_Class": predictions,
    "Default_Probability": default_probabilities
})

prediction_results.head()
````

### Business Interpretation

The trained LightGBM pipeline generates both a predicted credit risk classification and a probability of default for each loan applicant. While the predicted class provides a binary lending decision (Default or Non-Default), the probability score offers a more detailed assessment of the applicant's level of credit risk.

Financial institutions can use these probability estimates to prioritize high-risk applications for manual review, implement risk-based pricing strategies, optimize credit approval policies, and strengthen portfolio risk management. By combining automated machine learning predictions with business judgment, lenders can make faster, more consistent, and data-driven credit decisions.

## 12.5 Export Prediction Results

In this section, the generated credit default predictions are exported to a CSV file. The output file contains each customer's unique identifier, predicted credit risk class, and probability of default.

Exporting the prediction results enables the model outputs to be integrated into downstream business processes, such as loan approval systems, portfolio monitoring, risk reporting, or competition submissions. Saving the predictions also improves reproducibility and facilitates future analysis without rerunning the entire prediction pipeline.

````python
# ============================================================
# 12.5 Export Prediction Results
# ============================================================

# Create output directory if it does not exist
from pathlib import Path

output_dir = Path("../reports")
output_dir.mkdir(parents=True, exist_ok=True)

# Save prediction results
output_file = output_dir / "credit_default_predictions.csv"

prediction_results.to_csv(output_file, index=False)

print("=" * 60)
print("Prediction results exported successfully!")
print("=" * 60)

print(f"\nFile saved to: {output_file}")
print(f"Total Records Exported: {len(prediction_results):,}")
````

### Business Interpretation

Exporting the prediction results converts the model outputs into a usable business asset. The generated file can be integrated into loan processing systems, credit risk dashboards, portfolio monitoring frameworks, or regulatory reporting workflows.

Each prediction includes both the estimated probability of default and the corresponding risk classification, enabling financial institutions to support automated credit decisions, prioritize high-risk applications for further review, and maintain a consistent, data-driven lending process.
